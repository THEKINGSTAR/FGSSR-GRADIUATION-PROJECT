```python
import json

# Read the CSV file into a DataFrame
projects_df = pd.read_csv("projects_describtions.csv")

# Group data by ID and aggregate the values
combined_df = projects_df.groupby('ID').agg({
    'Project Title': 'first',
    'Requirements': ' '.join,
    'Domain': 'first',
    'Complexity Level': 'first',
    'Description': 'first'
})

# Format the data as a list of dictionaries
json_data_Complexity = []
for index, row in combined_df.iterrows():
    project_data = {
        'title': row['Project Title'],
        'Requirements': row['Requirements'],
        'Domain': row['Domain'],
        'Complexity Level': row['Complexity Level'],
        'Description': row['Description']
    }
    json_data_Complexity.append(project_data)

# Convert the data to JSON
json_data_Complexity = json.dumps(json_data_Complexity, indent=4)

# Save the JSON data to a file
with open('json_data_Complexity.json', 'w') as file:
    file.write(json_data_Complexity)
```
